```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import pickle
```

```python
df = pd.read_csv('./trip_tag.csv')
df = df.drop('_id', axis=1)
```

```python
df2 = df[['食物', '服務', '實惠', '氛圍']]
```

```python
df2 = df2.replace('None', np.nan)
df2['食物'] = pd.to_numeric(df2['食物'])
df2['服務'] = pd.to_numeric(df2['服務'])
df2['實惠'] = pd.to_numeric(df2['實惠'])
df2['氛圍'] = pd.to_numeric(df2['氛圍'])
df2 = df2.fillna(1)
df2
```

```python
df_list = df2.values.tolist()
df_list
```

```python
# 建 K-means 模型
X = np.array(df_list)               
model = KMeans(n_clusters=5, random_state=0).fit(X)
model.labels_
```

```python
model.cluster_centers_
```

```python
# 儲存 K-means 模型
with open("kmeansmodel.pkl", "wb") as f:
    pickle.dump(model, f)
```

```python
# 讀取 K-means 模型
with open("kmeansmodel.pkl", "rb") as f:
    model = pickle.load(f)
```
